## 5.1.1 Data Wrangling and Visualizing

**Exercise 5.1.1.1** Read in the dataset "nchs_death_causes.csv" as a list of lists. You will need to use the `csv` package and then will want to go through each row appending it to a list you create.

```python
import csv
with open('nchs_death_causes.csv', 'r', newline='') as death_file: 
    reader = csv.reader(death_file)
    data = list(reader)
    
data
```

**Exercise 5.1.1.2** Make a list that is a subset of the data, containing only one cause of death and one state.

```python
subset_list = []
#subset_list.append(data[2][3])
#subset_list.append(data[2][2])
#subset_list

for element in data:
    if element[3] == 'Alaska' and element[2] == 'Unintentional injuries':
        subset_list.append(element)
        
subset_list
```

**Exercise 5.1.1.3** Plot a line graph that shows the development of the death rate over time.

```python
import matplotlib.pyplot as plt

values = [int(element[4]) for element in subset_list]
labels = [int(element[0]) for element in subset_list]
labels.reverse () #denne vender årstallene om
#elements skal ændres fra string til integer 

plt.figure(figsize=(10,7), dpi=120)

plt.plot(values)

plt.title("Graph to show unintentional injuries in Alaska by year")
plt.xlabel("Year")
plt.ylabel("Deaths per 100,000")
plt.xticks(range(0,len(labels)), labels) #, rotation=90)

plt.show()
```

**Exercise 5.1.1.4** Make your figure self-contained and export it/save it to your machine.

```python
import os

values = [int(element[4]) for element in subset_list]
labels = [int(element[0]) for element in subset_list]
labels.reverse () #denne vender årstallene om
#elements skal ændres fra string til integer 

plt.figure(figsize=(10,7), dpi=120)

plt.plot(values)

plt.title("Graph to show unintentional injuries in Alaska by year")
plt.xlabel("Year")
plt.ylabel("Deaths per 100,000")
plt.xticks(range(0,len(labels)), labels) #, rotation=90)

file_dir = os.getcwd()

plt.savefig(file_dir + "/Alaska_injuries.png")
```

**Exercise 5.1.1.5** Make a list that is a subset of the data, containing _two_ causes of death that might be interesting to compare for one state.

```python
two_subset_list = []
two_subset_list.append(subset_list)

for element in data:
    if element[2] == "Heart disease":
        two_subset_list.append(element)
        
two_subset_list
```

**Exercise 5.1.1.6** Create a _self-contained_ scatter plot that compares the two causes of death in the state.

```python
values_injury = [int(element[4]) for element in two_subset_list[0]]
values_heart = [int(element[4]) for element in two_subset_list[1]]

plt.figure(figsize=(10,7), dpi=120)


plt.title("Scatter plot showing heart disease and unintentional injuries in Alaska")
plt.xlabel("Unintentional injuries")
plt.ylabel("Heart deasease")


plt.plot(values_injury, values_heart, 'o')
```
